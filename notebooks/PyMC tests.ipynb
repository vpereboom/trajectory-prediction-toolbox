```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')

# Initialize random number generator
np.random.seed(123)

# True parameter values
alpha, sigma = 1, 1
beta = [1, 1.2]

# Size of dataset
s = 1000

# Predictor variable
xd1 = pm.distributions.continuous.SkewNormal.dist(mu=2.0,tau=0.8, alpha=1)
X1 = np.random.randn(s) #xd1.random(size=s)
X2 = 3 + np.random.randn(s) * 1.2

# Simulate outcome variable
Y = alpha + beta[0]*X1 + beta[1]*X2# + np.random.randn(size)*sigma
x = X1 + X2 - (X1*X2)

plt.figure(figsize=(10,10))
plt.scatter(X1, Y)
plt.scatter(X2, Y)
plt.scatter(x, Y)
plt.show()

```
